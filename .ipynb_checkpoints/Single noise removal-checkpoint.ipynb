```python
import cv2
import os
import tensorflow as tf

# Path to the directory containing the Herlev dataset
dataset_dir = 'Herlev dataset/train/carcinoma_in_situ'

# List all image files in the dataset directory
image_files = [os.path.join(dataset_dir, filename) for filename in os.listdir(dataset_dir) if filename.endswith('.bmp')]

# Apply median filtering to each image and preprocess using TensorFlow
for image_file in image_files:
    # Load the image using OpenCV
    image = cv2.imread(image_file)

    # Convert the image to RGB (OpenCV reads images in BGR format)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply median filtering
    filtered_image = cv2.medianBlur(image_rgb, 3)  # Change the kernel size (3) as needed

    # Display the original and filtered images
    cv2.imshow("Original Image", image_rgb)
    cv2.imshow("Filtered Image", filtered_image)
    cv2.waitKey(0)

# Close all open windows
cv2.destroyAllWindows()
```
